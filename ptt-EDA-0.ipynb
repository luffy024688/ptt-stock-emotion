```python
import pandas as pd
df = pd.read_csv('/Users/yichinghau/Desktop/AI-course/ptt/data/pttstock.csv',encoding='big5hkscs')
```

```python
df.columns = ['count','title', 'link','date','content','push']
df['Date'] = pd.to_datetime(df['date'],format = '%a %b %d %H:%M:%S %Y')
df['Date'] = df['Date'].apply(lambda x:x.date())
df = df[['count','Date','title']]#[['count','Date','title','push','content']]
```

```python
#df[df['Date'].isna()==True]
```

```python
## stop word ------sw 優化
f = open('./dict/stopword_ch.txt','r')
sw = f.readlines()
f.close()

from opencc import OpenCC
cc = OpenCC('s2t')  # convert from Simplified Chinese to Traditional Chinese
sw = [cc.convert(w).replace('\n','') for w in sw]
```

```python
import jieba
df_new = pd.melt(df, id_vars=["count",'Date']) #R-gather
df_new['a'] = df_new['value'].apply(lambda x: list(jieba.cut(x)))
df_new['b'] = df_new['a'].apply(lambda x:[w for w in x if w not in sw])
df_new['c'] = df_new['b'].apply(lambda x:' '.join(x))
```

```python
df_new[['Date','c']].rename(columns={'Date':'created_time','c':'title'}).to_csv('./data/stock_comments_seg_us.csv',index=False)
```

```python
for i in df_new['a']:
    print(i)
```

```python
### Model
## bag of words
from sklearn.feature_extraction.text import CountVectorizer
model = CountVectorizer()
model.fit(df_new['c'])
X = model.transform(df_new['c'])
model.vocabulary_
```

```python
model3 = CountVectorizer(list(df_new['c']),stop_words=sw,min_df=0.005,max_df=0.95)# min_df=0.25表最少25%doc，min_df=2表最少出現2doc
Y = model3.fit_transform(list(df_new['c']))
len(model3.vocabulary_)
model3.vocabulary_
Y.toarray()
```

```python
### TF-IDF- (njobs???)
from sklearn.feature_extraction.text import TfidfTransformer
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(X)
tfidf_weight = tfidf.toarray()  # shape = len(Document)*len(word) 
word = model.get_feature_names()


for i in range(len(tfidf_weight)):
    print("-------output {}-th document tf-idf weight------".format(i))
    for j in range(len(word)):
        if (tfidf_weight[i,j]>0):
            print(word[j],tfidf_weight[i,j])
```

```python
df_per_date = df_new.groupby('Date')['c'].sum().reset_index().sort_values('Date')
```

```python
### 情緒模型 研究
##https://www.slideshare.net/ssuser968e8d/basic-concept-of-sentiment-analysis-and-affective-computing
##https://github.com/UDICatNCHU/Swinger
from Swinger import Swinger
s = Swinger()
s.load('LogisticRegression') # 或是其他模型例如MultinomialNB
df_per_date['output'] = df_per_date['c'].apply(lambda x :s.swing(x))
```

```python
df_stock = yf.download('0050.TW',start = '2020-01-04').reset_index()
df_stock.columns = ['date','open', 'high', 'low', 'close', 'Adj Close', 'volume']
df_stock['close-1']  = df_stock['close'].shift(1)
df_stock['dif'] = df_stock['close']-df_stock['close-1']
df_stock.to_csv('./data/sh000001_us.csv',index=False)
```

```python
df_per_date['Date'] = df_per_date.Date.astype(str)
df_stock['date'] = df_stock.date.astype(str)
df_all = df_per_date.merge(df_stock,left_on = 'Date',right_on='date',how='left')[['Date','c','output','dif']]
```

```python
df_all['dif2'] = df_all['dif'].fillna(method='ffill')
```

```python
df_all.to_csv('./data/df_merge.csv',index=False)
```
